### 데이터 조작

- feature engineering

if 0 ~ 1 까지는 1로, 1 ~ 2 은 2 , 2~3 은 3 으로 새로운 구간을 만들고 싶어.

나이같은 것들, 일일이 사람의 나이로 분석하는 것보다 각 나이대로 구분해서 분석하는 것이 의미가 있을 때가 있어.



```python
csData["PRODUCTAGE_NEW"] = np.where(csData.PRODUCTAGE < 1 ,1 ,0) #where(조건, 참일 경우 결과, 거짓일 경우 결과)
# csData.loc[csData.PRODUCTAGE < 1, "PRODUCTAGE_NEW"] # 이렇게도 가능하지만 속도가 매우 느림
csData
```

```python
# 데이터를 생성했으면 검증이 필요해
csData.loc[(csData.PRODUCTAGE_NEW != 1) & (csData.PRODUCTAGE < 1) ]
```

```python
# if 0 ~ 1 까지는 1로, 1 ~ 2 은 2 , 2~3 은 3 으로 새로운 구간을 만들고 싶어.
csData["PRODUCTAGE_NEW"] = np.where(csData.PRODUCTAGE < 1 ,1 ,
          np.where(csData.PRODUCTAGE <2 ,2 ,3))
csData
```

```python
def ProductAge_fillter( inProductAge ):

    #inProductAge = 2
    outProductAge = 0

    # ProductAge에 따른 NEW 값 분리
    if inProductAge < 1 :
        outProductAge = 1
    elif inProductAge < 2:
        outProductAge = 2
    else :
        outProductAge = 3

    return outProductAge
```

```python
csData.PRODUCTAGE_NEW = csData.PRODUCTAGE.apply(ProductAge_fillter) # 특정 칼럼에 조건을 만들 함수 적용
csData
```

```python
def trimmedAvg( inList) :
    """
    description : calc mean except min, max Value
    """

    minValue = min(inList)
    maxValue = max(inList)
    inList.remove(minValue)
    inList.remove(maxValue)

    trimmed_Mean = sum(inList) / len(inList)
    return trimmed_Mean
```

```python
inList = [100,120,130,150,170]
trimmedAvg(inList)
```

## 이동평균 구하기

자신을 기준으로 이동하면서 편균을 구하는 것 // ***그 주변 기간의 추세를 알기 위해 필요해***

- 표준편차 등 꼭 평균이 아니어도 활용할 수 있어.


++++ 상한 하한 안에 데이터가 있나 없나 , 이상치를 정제하는 느낌으로도 활용할 수 있지.

#### 코드


```python
maData = pd.read_csv("../dataset/kopo_product_volume.csv")
maData
```

```python
# 정렬을 안했을 때 이동평균 // 사실 의미가 없잖아. 기간별도 아니고, 암것도 아니니.
maData.VOLUME.rolling(window = 5, center = True, min_periods= 3).mean()
```

```python
#데이터 정렬하기
sortKey = ["REGIONID","PRODUCTGROUP", "YEARWEEK"]
maData = maData.sort_values(sortKey)
```

```python
# 이동평균 구하기
maData["MA5"] = maData.VOLUME.rolling(window = 5, center = True, min_periods= 3).mean()
# window : 나를 기준 범위, center : 중앙에 둘지 여부, min_periods : 최소값으로 NaN 채우기
#min_periods : 계산할 최소 크기(기간) 입니다. window 안의 값의 수가 min_periods의 값보다 작을경우 NaN을 출력합니다.
maData
```

단 이렇게 했을 때 지역이 달라지는 부분에서는 이동평균의 의미가 없어져.

그룹이 달라졌을 때, 그 부분이 이상한 값이 돼.


그래서 추가적으호 groupby로 rolling 을 넣어줘. (이건 심화과정에서)


```python
# 추가적으로 이렇게 하면 추세가 어떻게 되는지 알 수 있지.
maData["MA5"] = maData.VOLUME.rolling(window = 5, center = True, min_periods= 3).mean()
maData["STD5"] = maData.VOLUME.rolling(window = 5, center = True, min_periods= 3).std()
maData["UPPERBAND"] = maData.MA5 + maData.STD5
maData["LOWERBAND"] = maData.MA5 - maData.STD5
maData["SEASON"] = maData.VOLUME/ maData.MA5

maData
# groupby 대안 쿼터를 나눠서 분석 할 수 있어. 궁금하면 유튜브를 찾아보자.
```

### Join에 대해서

```python
joinMaster = pd.read_csv("../dataset/join_region_master.csv",encoding = "ms949") ## 한글 오류일 시 encoing 을 변경해줘.
joinSellout = pd.read_csv("../dataset/join_kopo_product_volume.csv")
print(joinMaster,end = "\n\n")
print(joinSellout)
```

```python
# REGIONID로 JOIN 하기 // pd.merge(left : dataset1 , right : dataset2 ,left_on : leftkey, right_on : rightkey, how : join method)
# how로 innerjoin을 많이 쓰는데 실무에선 left Join을 써야해.
#innerJoin 하면 데이터를 잃어버릴 수 있어.

print("inner Join ")
print(pd.merge(left = joinMaster, right = joinSellout, left_on = "REGIONID", right_on = "REGIONID", how = "inner")) # 외래키가 같으면 그냥 on으로 해도 괜찮음.
print("Left Join ")
print(pd.merge(left = joinMaster, right = joinSellout, left_on = "REGIONID", right_on = "REGIONID", how = "left"))
print("right Join ")
print(pd.merge(left = joinMaster, right = joinSellout, left_on = "REGIONID", right_on = "REGIONID", how = "right"))
```

```python
# 값만 같으면 되서 이런 오류도 있어.
joinMaster2 = joinMaster.rename(columns = {"REGIONID" : "SALESID"})
joinMaster2
```

```python
pd.merge(left = joinMaster2, right = joinSellout, left_on = "SALESID", right_on = "REGIONID", how = "left")
```

```python
targetRegion = ["A01","A03"]

joinMaster3 = joinMaster.loc[joinMaster.REGIONID.isin(targetRegion)]
joinMaster3
```

```python
pd.merge(left = joinMaster3, right = joinSellout, left_on = "REGIONID", right_on = "REGIONID", how = "right")
```
